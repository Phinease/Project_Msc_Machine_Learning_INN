```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def load_monte_carlo_reflectances(path):
    df_montecarlo = pd.read_csv(path, usecols=range(1,152), dtype = np.float64)
    column_names = np.arange(400, 702, 2)
    df_montecarlo.columns = column_names
    return df_montecarlo

def load_monte_carlo_parameters(path):
    df_montecarlo = pd.read_csv(path)
    return df_montecarlo
```

```python
reflectance_path = '/Users/conor/Documents/HVS/hvs_research_and_development/monte_carlo/mcx_dgx/mcx_parallel/parameter_series/mc_reflectance.csv'
parameter_path = '/Users/conor/Documents/HVS/hvs_research_and_development/monte_carlo/mcx_dgx/mcx_parallel/parameter_series/mc_parameters.csv'

mc_reflectances = load_monte_carlo_reflectances(reflectance_path)
mc_parameters = load_monte_carlo_parameters(parameter_path)
mc_parameters
```

```python
mc_reflectances
```

```python
def display_parameter_series(wavelengths, parameter_values, min_val, max_val, param = 'StO$_2$'):
    ax = plt.axes()
    ax.set_prop_cycle('color',[plt.cm.seismic(i) for i in np.linspace(0, 1, len(parameter_values))])
    plt.plot(wavelengths, parameter_values.T)
    sm = plt.cm.ScalarMappable(cmap=plt.cm.seismic, norm=plt.Normalize(vmin = min_val, vmax = max_val))
    plt.colorbar(sm).set_label(param, labelpad = 20, rotation=270)
    plt.title('Effect of Varying ' + param)
    plt.xlabel("Wavelength (nm)")
    plt.ylabel("Reflectance (a.u.)")
    plt.xlim([wavelengths[0], wavelengths[-1]])
    plt.ylim([0,1])
    plt.show();
```

```python
start_idx = 0
end_idx = 14
min_val = mc_parameters['StO2_2'][start_idx]
max_val = mc_parameters['StO2_2'][end_idx]
display_parameter_series(mc_reflectances.columns, np.array(mc_reflectances.iloc[start_idx:end_idx,:]), min_val, max_val, param = 'Dermis StO$_2$')
```
